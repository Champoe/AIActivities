try natin pag pilipinas, tignan natin kung may available na data sa kanya. tignan natin mga available na data bawat country

```python
ph_data = df[df.country == "Philippines"]
ph_data
```

try natin hanapin kung anong pinakamataas na population na inilaki ng pinas based sa data

```python
def largest_population_gain(years, populations):
    max_gain = 0
    max_gain_year = None

    for i in range(1, len(populations)):
        gain = populations[i] - populations[i - 1]
        if gain > max_gain:
            max_gain = gain
            max_gain_year = years[i]

    return max_gain_year, max_gain

years = ph_data['year'].values
populations = ph_data['population'].values
max_gain_year, max_gain = largest_population_gain(years, populations)

print(f"The Year where the Philippines gained the most is the year {max_gain_year} with {max_gain} na katao.")
```

### ngayon, graph natin ang data ng pinas

```python
sns.lineplot(x = ph_data['year'] , y = ph_data['population']) 
plt.xlabel("Year",fontsize=20, fontweight='bold')
plt.ylabel("Population",fontsize=20, fontweight='bold')
plt.title("Philippines' Population Every 5 Years", fontsize=15, fontweight='bold')
plt.show()
```
